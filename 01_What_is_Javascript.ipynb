# 1장 자바스크립트란 무엇인가
## 1.1 간추린 역사
* 웹이 인기를 얻으며 클라이언트 스크립트 언어에 대한 요구가 커져감
    * 당시 인터넷 속도: 28.8kbps
    * 단순한 유효성 검사를 위해 매번 서버까지 왕복
* 넷스케이프에서 일하던 '브랜든 아이크' Mocha, 나중에 LiveScript 라고 불리는 스크립트 언어 개발
    * 1995년 넷스케이프 내비게이터2의 출시에 맞춰 브라우저와 서버 모두 사용하려는 의도
    * 출시 직전 자바스크립트로 이름을 변경
        * 당시 자바가 얻고 있던 인기에 편승하기 위해
        * 자바스크립트와 자바는 전혀 상관없음
* 자바스크립트 1.0 대성공
* 넷스케이프 내비게이터3에 자바스크립트 버전 1.1 포함
    * 넷스케이프는 웹을 선도하는 회사로 자리매김
* 그 즈음(1996년 8월) 마이크로소프트에서 인터넷 익스플로러 3 출시
    * 여기에 포함된 자바스크립트는 JScript라 부름
    * 넷스케이프와 저작권 문제를 피하기 위함
* 마이크로소프트가 독자적으로 자바스크립트를 구현함에 따라 넷스케이프 내비게이터의 자바스크립트와 인터넷 익스플로러의 JScipt 두 가지 자바스크립트 버전이 존재
    * C언어나 기타 프로그래밍 언어와 달리 문법이나 기능을 제어할 표준이 없었고, 각기 다른 버전은 혼란을 가중시킴
* 자바스크립트를 표준화 하기로 결정
* 1997년 자바스크립트 1.1 유럽 컴퓨터 제작자 협회(European Computer Manufacturers Association:ECMA)에 제안
    * '문법과 의미를 표준화하여 일반적인 목적에 쓸 수 있고 플랫폼을 가리지 않으며 제조사에 중립인 스크립트 언어'를 만들기로 함
    * 몇 달에 걸쳐 자바스크립트 1.1을 손질해 ECMA-262(기술 규격)를 만듦
    * 이 표준은 ECMAScript(보통 에크마스크립트로 발음)라는 이름의 스크립트 언어를 정의
* 이듬해 국제 표준화 기구(ISO)와 국제 전기 표준회의(IEC)에서도 ECMAScript를 표준(ISO/IEC-16262)으로 받아들임
    * 이때부터 브라우저 제작자들은 ECMAScript를 자사 자바스크립트 구현의 기반으로 삼으려 했으나 완성도는 제각기 달랐음
    
## 1.2 자바스크립트 구현
* 자바스크립트와 ECMAScript를 비슷한 말로 여기는 사람들이 많지만 자바스크립트는 ECMA-262에 정의된 것 보다 훨씬 많은 내용을 담고 있음
    * 자바스크립트 구현
        * 코어(ECMAScript)
        * 문서 객체 모델(DOM)
        * 브라우저 객체 모델(BOM)
        
### 1.2.1 ECMAScript
* ECMAScript 언어는 웹 브라우저에 종속되지 않음
* 사실 ECMAScript 언어에는 입력이나 출력을 담당하는 메서드가 전혀 없다.
* 더 견고한 스크립트 언어를 만들기 위한 기반 언어
* 웹 브라우저는 ECMAScript를 구현하는 '호스트 환경'일 뿐
* 호스트 환경은 ECMAScript를 기본적으로 구현하고 이 환경 자체의 인터페이스 구실을 하도록 디자인된 확장을 제공
    * 확장이란 문서 객체 모델(BOM)같은 것을 말함
    * 이러한 확장은 ECMAScript의 핵심 타입과 문법을 이용해 환경에 밀접한 기능을 제공
* ECMA-262는 다음과 같은 언어의 저수준에 해당하는 부분을 정의
    * 문법
    * 타입
    * 선언문
    * 키워드
    * 예약어
    * 연산자
    * 객체
* ECMAScript는 단순히 명세에서 설명하는 부분들을 구현한 언어를 부르는 말
    * 자바스크립트는 ECMAScript를 구현한 것이며 어도비 액션스크립트도 마찬가지
    
#### ECMAScript '판'
* ECMAScript의 여러 버전을 '판'이라 부르는데 이는 ECMA-262의 개별 구현을 가리킴

![1_1.png](1_1.png)

#### ECMA 표준 준수란 무슨 의미인가
* ECMA-262에는 ECMAScript표준 준수에 관한 정의도 있다.
    * 다음 목록을 모두 만족해야만 ECMAScript를 구현한 것으로 간주
        * 모든 타입과 값, 객체, 프로퍼티, 함수, 프로그램 문법과 시맨틱을 ECMA-262에 설명한 그대로 구현해야 합니다.
        * 유니코드 문자 표준을 지원해야 합니다.
    * 다음 목록은 제작자의 재량에 따라 해도 되고 하지 않아도 된다.
        * ECMA-262에서 정의하지 않은 타입이나 값, 객체, 프로퍼티, 함수를 추가 할 수 있다.
        * ECMA-262에 정의되지 않은 프로그램이나 정규 표현식 문법을 지원할 수 있습니다.
            * 내장된 정규 표현식을 확장하거나 변형해도 된다는 의미
    * 이를 통해 개발자들이 ECMAScript에 기초해 얼마든지 새로운 언어를 구현할 수 있는데, 이 역시 ECMAScript가 인기 있는 언거가 된 이유 중 하나

### 1.2.2 문서 객체 모델
* '문서 객체 모델' (Document Object Model: DOM)
    * XML을 HTML에서 사용할 수 있도록 확장한 API
    * DOM은 전체 페이지를 노드의 계층 구조로 변환
    * HTML(XML) 페이지의 각 부분은 각기 다른 데이터를 포함하는 다양한 타입의 노드로 표현
    
        ```html
        <html>
            <head>
                <title>Sample Page</title>
            </head>
            <body>
                <p>Hello World!</p>
            </body>
        </html>
        ```

    * 이 코드는 DOM을 통해 노드의 계층 구조인 다이어그램으로 나타낼 수 있음
    
    ![1_3.png](1_3.png)
    
    * DOM은 문서를 표현하는 트리를 생성하고 개발자는 이를 통해 문서의 콘텐츠와 구조를 자유롭게 수정 할 수 있게 됨
    * DOM API를 통해 노드를 쉽게 제거, 추가, 교체, 수정할 수 있음
    
#### DOM이 필요한 이유
* 익스플로러4와 넷스케이프 내비게이터4가 각기 다른 동적 HTML(DHTML)을 지원하면서 개발자들은 처음으로 웹 페이지를 새로고침하지 않고도 콘텐츠의 모양을 바꿀 수 있게 되었음
    * 서로 개발한 DHTML은 서로 완전히 달랐음, 재앙
* 웹 통신 표준을 만들던 월드 와이드 웹 컨소시엄(W3C)에서 DOM관련 작업을 시작

#### DOM레벨
* DOM 레벨 1
    * 1998년 10월 W3C 권고
    * 두가지 모듈로 구성
        * DOM 코어
            * XML 기반 문서를 노드 트리로 변환해 문서의 어느 부분이든 쉽게 접근하고 조작
        * DOM HTML
            * HTML에 밀접한 객체와 메서드를 DOM코어에 추가 확장
* DOM 레벨 2
    * 마우스 및 사용자 인터페이스 이벤트 지원
    * 범위 이동 추가
    * 객체 인터페이스를 통해 캐스케이딩 스타일 시트(CSS) 지원
    * DOM 코어
        * XML 네임스페이스 지원
    * 새로운 모듈 도입
        * DOM Views: 문서의 다양한 뷰, 예) CSS 스타일 적용 전후 추적 인터페이스
        * DOM 이벤트: 이벤트와 이벤트 처리에 관한 인터페이스
        * DOM 스타일: CSS를 통해 요소의 스타일을 바꾸는 인터페이스
        * DOM 이동과 범위: 문서 트리를 이동하고 조작하는 인터페이스
* DOM 레벨 3
    * 문서를 저장하고 불러오는 통일된 방법
    * 문서가 유효한지 검사하는 방법
    * DOM 코어
        * XML 인포셋, XPath, XML 베이스를 포함 XML 1.0 전체를 지원하도록 확장

#### 기타 DOM
* 다른 언어에서 독자적으로 만든 DOM 표준 존재
    * XML 기반, 각각 고유한 메서드와 인터페이스를 DOM 에 추가
        * 크기 조절이 가능한 벡터 그래픽(Scalable Vector Graphic: SVG 1.0)
        * 수학 마크업 언어(Mathematical Markup Language: MathML 1.0)
        * 동기화된 멀티미디어 확장 언어(Synchronized Multimedia Integation Languiage: SMIL)

### 1.2.3 브라우저 객체 모델
* 브라우저 창에 접근하고 조작할 수 있게하는 인터페이스
* 별개의 컨텍스트에서 브라우저와 상호작용 할 수 있음
* BOM을 주관하는 자바스크립트 표준이 존재하지 않음
* 브라우저 창과 프레임을 다루는 인터페이스
* 일반적으로 브라우저에 고유한 자바스크립트 확장도 BOM의 일부로 간주
    * 브라우저 창을 새로 띄움
    * 브라우저 창을 움직이거나 크기를 조절하고 닫음
    * 브라우저의 상세 정보를 반환하는 navigator 객체
    * 불러온 상세 정보를 반환하는 navigator 객체
    * 불러온 페이지의 상세 정보를 반환하는 location 객체
    * 사용자의 화면 해상도에 대해 상세 정보를 반환하는 screen 객체
    * 쿠키 지원
    * XMLHttprequest와 인터넷 익스플로러의 ActiveXObject 같은 커스텀 객체
    

## 1.3 자바스크립트 버전
* 모질라
    * 넷스케이프를 계승, 자바스크립트 버전 규칙을 쓰는 유일한 브라우저
* 넷스케이프/모질라 브라우저의 자바스크립트 버전
    * 이 버전 규칙은 파이어폭스 4에서 자바스크립트 2.0을 도입할 것이라는 예상하에 세워진 것
    * 자바스크립트의 급격한 발전 때문에 이 계획은 더이상 사용할 수 없게 되었음
    



## 1.4 요약
* 자바스크립트는 웹 페이지와 상호작용하도록 디자인된 스크립트 언어이며 다음 세 부분으로 구성
    * ECMA-262에서 정의하는 ECMAScript
        * 핵심 기능을 제공
    * 문서 객체 모델(DOM)
        * 웹 페이지 콘텐츠를 조작하는 메서드와 인터페이스를 제공
    * 브라우저 객체 모델(BOM)
        * 브라우저와 상호작용하는 메서드와 인터페이스를 제공
* 이들 부분에 대한 주요 웹 브라우저의 지원은 각기 다르다
* ECMAScript 3에 대한 지원은 일반적으로 좋은 편이며, ECMAScript 5에 대한 지원은 점점 나아지는 중, DOM지원은 편차가 심함
* BOM은 최근에 HTML5에서 표준화 되고 있으므로 브라우저 사이에 일부 공통인 점이 있긴 하지만 매우 다르다고 생각해야 함.
