# 3장 언어의기초
    
## 3.1 문법
* ECMASCript의 문법은 대부분 C언어, 자바 등 C와 비슷한 언어에서 차용한것
    * 대소문자구분: test와 Test는 다른변수
        * 식별자: 변수나 함수, 프로퍼티, 함수 매개변수의 이름

        * 글자나 밑줄, 달러, 기호, 숫자를 자유롭게 쓸 수 있다.
        * ECMAScript 식별자는 카멜케이스를 쓴다.(이 형식을 따르길 권함) ex) firstSecond, myCar 등

    * 주석:  
        ```javascript
        // 한 줄 주석
        ```
        
        ```javascript
        /* 
        * 여러줄 주석은
        * 이런식으로 쓰는거 다 아시죠 
        */
        ```
    * 스트릭트 모드: ECMAScript 5에서 '스트릭트 모드' 개념 도입함. 
         * 스트릭트 모드는 기존과는 다른 방식으로 자바스크립트를 파싱하고 실행하라고 지시하는것
         * 전체 스크립트에 스트릭트 모드를 적용하려면 "use strict" 를 맨 위에 추가한다. 
         * 스트릭트 모드를 지원하는 자바스트립트 엔진이 이를 인식하고 스트릭트 모드로 전환함
         * 함수 하나만 스트릭트 모드로 실행할 수 있다. 
        ```javascript
        function doSomething() {
            "use strict"
            // 함수 본문
        }
        ```
            [스트릭트모드](https://developer.mozilla.org/ko/docs/Web/JavaScript/Reference/Strict_mode/Transitioning_to_strict_mode)

    * 문장: ECMAScript에서는 각 문장은 세미콜론으로 종료한다. 항상 세미콜론을 쓰길 권장한다. 
        * 타이핑하다가 멈췄던 부분을 깜박 잊는 등의 에러를 쉽게 발견 할 수 있고
        * 여분의 공백을 제거해서 코드를 압축할 수도 있는데 세미콜론이 없으면 압축했을때 문법에러가 발생할 수 있음.
        
        ```javascript
        var sum = a + b // 세미콜론 없어도 유효하지만 권장 X
        var sum = a - b; // 권장
        ```  
    * 중괄호(코드 블록)사용 권장
    ```javascript
        if (test){
            alert(test);
        }
     ```  
     
## 3.2 키워드와 예약어
* 예약어란, 자바스크립트 문법을 규정짓기 위한 특수 키워드를 말합니다. <br> 예약어로 지정된 키워드들은 식별자나 프로퍼티 이름으로 사용하지 않으며, 미리 정의된 전역 변수와 전역 함수 역시 사용하지 않는 편이 좋습니다. <br> 사용하더라도 오류는 발생하지 않지만, 본래의 기능을 사용할 수 없게 됩니다. <br> 또한 Window 객체의 이름과 DOM에서 사용하는 객체 이름도 사용하지 않습니다.

## 3.3 변수

* ECMAScript는 어떤 타입의 데이터라도 저장할 수 있는 느슨한 변수타입을 사용한다.
    * var message = "hi!;" -> message 변수가 hi 값을 저장했지만 변수에 문자열 타입을 지정한건 아님 즉, 값을 바꿀 수 있다.",  
    * var message "hi"; message = 100; // 유효하지만 권장X

* 지역변수
   ```javascript
    function test() {
        var message = "hi"; //지역변수
    }  
    test();
    alert(message); //에러

    ``` 
* 전역변수: 지역변수의 var 연산자를 제거하면서 message가 전역변수가 되었음
    ```javascript
    *function test() {
        message = "hi"; // 전역변수 (가능하지만 권장하지 않는 문법이라고함 -> 전역변수를 로컬에서 정의하면 관리어렵, 
                           var 키워드를 의도적으로 생략한것인지 혼란스러워서)
    }
    test();
    alert(message); // "hi"
    ```
    
* 변수 여러개 선언시 
    * 주의사항: strict mode 에서 변수 이름에 eval이나 arguments를 사용할 수 없다. (에러)
    ```javascript
    var message = "hi",
    found = false, 
    age = 29;
    ```
    
## 3.4 데이터 타입
   * 5가지의 기본적인 데이터 타입: undefined, Null, boolean, 숫자, 문자열 + 객체(이름-값 쌍의 순서 없는 목록)
      * typeof 연산자: 변수의 데이터 타입을 알아 낼때 값(변수)에 적용
      * 정의되지 않은 변수: undefined 
        ```javascript
        var message; // 변수 메세지를 선언했지만 초기화 하지 않았음
        alert(message == undefined); // true
        ```
        * 불리언: "boolean은 true와 false 두 가지 리터럴 값만 가진다 (대소문자 구분)
        ```javascript
            var found = true; 
            var lost = false;
        ```
        * 문자열: "string",
            * toString() 메서드는 숫자와 불리언, 객체, 문자열 값에 존재한다. (null 과 undefined에는 존재하지 않는다)
                ```javascript
                var age = 11;
                var ageAsString = age.toString(); // 문자열 "11"
                var found = true;
                var foundAsString = found.toString(); //문자열 'true'
                ```
            * 진법 매개변수를 통해 2,8,16진법 형식을 반환할 수 있다.
                매개변수를 쓰지 않으면 기본 값으로 10진법을 사용한다.
                ```javascript
                var num = 10; 
                alert(num.toString()); //"10" 
                alert(num.toString(2)); // "1010"
                alert(num.toString(8)); // "12" 
                alert(num.toString(10)); // "10"
                alert(num.toString(16)); // "a"
                ```
            * 값이 null, undefined 일때
                ```javascript
                var value1 = 10;
                var value2 = true;
                var value3 = null;
                var value4; 

                alert(String(value1)); // "10"
                alert(String(value2)); // "true"
                alert(String(value3)); // "null"
                alert(String(value4)); // "undefined"
                ```

        * 숫자: "number"
            * 가장 기본적인 숫자 리터럴 형식은 10진법
            * 정수는 8진수 리터럴이나 16진수 리터럴로 나타낼 수도 있다. (첫번째 숫자 반드시 0, 그 뒤에 숫자는 0~7까지만 허용)
                ```javascript
                 var octalNum1 = 070; // 8진법 으로 쓴 56
                 var octalNum2 = 079; // 9가 범위를 벗어났으며 79로 간주
                 var octalNum3 = 08; // 8이 범위를 벗어났으며 8로 간주
                ```
            * 스트릭모드에서 8진법 리터럴 허용 안함
            * 16진수 리터럴은 첫 두 문자에 반드시 0x(대소문자 안가림)를 쓰고 그 뒤에 16진수 숫자(0-9,A-F)를 쓴다.
                ```javascript
                 var hexNum1 = 0xA; //10의 16진수 표기
                 var hexNum2 = 0x1f; //31의 16진수 표기
                ```
            * 실제 계산은 항상 10진수로 변환하여 계산한다.
            * 부동소수점 숫자 -> 정수보다 메모리를 두배로 소모
                ```javascript
                var floatNum1 = 1.1 ;
                var floatNum2 = .1 //유효하지만 권장 안함
                ```
                * 지수표기법
                      *대단히 크거나 작은 부동소수점 숫자 표현할때 e 앞의 숫자에 10을 e 뒤의 숫자만큼 곱한다.
                        ```javascript
                        var floatNum = 3.125e7; // 31,250,000 (3.125dp 10^7를 곱한다)
                        ```
                * isFinite()함수: 어떤값이 유한한지 최솟값과 최댓값 사이에 있는지 확인할때 사용(있으면 true)
            * NaN(Not a Number)
                * 의도한 조작이 실패했을 때 반환하는값으로 오류가 아님
                    ```javascript
                    alert(NaN == NaN); // false - NaN어떠한 값과도 일치하지 않음
                    alert(isNaN(NaN)); // true - 숫자가 아닌값인지 검사
                    alert(isNaN(10)); // false - 10은 숫자
                    alert(isNaN("blue")); // true - 숫자로 바꿀 수 없다
                    alert(isNaN(true)); // false - 불리언 값은 숫자 1로 바꿀 수 있다.
                    ```
            * 숫자변환
                * 숫자가 아닌 값을 숫자로 바꾸는 함수 3가지 - Number(), parseInt(), parseFloat()
                 * Number(): 어떤 데이터 타입에든 쓸 수 있다. 다음과 같은 규칙에 기반하여 변환한다
                    ```javascript
                    var num1 = Number("Hello"); // 숫자형 값이 전혀 없으므로 NaN
                    var num2 = Number(""); // 빈 문자열은 0
                    var num3 = Number("000011"); // 리딩제로를 무시하고 11 
                    var num4 = Number(true); //1 
                    ```
                 * parseInt(), parseFloat(): 문자열을 숫자로 바꾸는 함수
                 
                    * paseInt() 함수 예시
                    ```javascript
                    var num1 = parseInt("1234blue"); //1234
                    var num2 = parseInt(""); // NaN
                    var num3 = parseInt("0xA"); // 16진수 10
                    var num4 = parseInt(22.5); // 22
                    var num5 = parseInt("70"); // 70
                    var num6 = paseInt("0xf"); // 16진수 15
                    var num7 = paseInt("0xAF",16); // 175 두번째 매개변수에서 16진수 임을 명기
                    var num8 = paseInt("AF",16); // 175
                    var num9 = paseInt("AF"); // NaN
                    var num10 = parseInt("10",2); // 2진수 10은 10진수 2
                    var num11 = paseInt("10",8); // 8진수 10은 10진수 8
                        ```
                    * parseFloat() 예시
                    ```javascript
                    var num1 = parseFloat("1234blue"); // 1234
                    var num2 = parseFloat("0xA"); // 0
                    var num3 = parseFloat("22.5"); // 22.5
                    var num4 = parseFloat("0908.5"); // 908.5
                    var num5 = parseFloat("3.125e7"); // 31240000
                        ```

        * 함수를 제외한 객체 또는 null:'object'
            * null 은 빈 객체를 가리키는 포인터이므로 null typeof를 호출하면 object를 반환합니다. 
            ```javascript
            var car = null;
            alert (typeof car); //object
            ```
            * 변수 정의할때 해당 변수가 객체를 가리키게 할 생각이면 null로 초기화 하기
                ```javascript
                if (car != null) {
                    //car를 사용하는 코드
                }
                ```
                * undefined 는 null에서 파생됨
                ```javascript
                alert(null == undefined ); //true
                ```

        * 함수: "function"
            ```javascript
            var message = "some string",
            alert(typeof message); // 'string'
            alert(typeof (message)); // 'string'
            alert(typeof 95); // 'number'
            ```

## 3.5 객체타입
  * ECMAScript에서 객체는 데이터와 기능의 집합이다. 객체는 new 연산자 다음에 새로 만들 객체 타입의 이름을 쓴다 
        ```javascript
        var o = new Object(); // 괄호 생략해도 되는데 쓰는것을 권장
        ```
        
    * Object의 인스턴스는 다음 프로퍼티와 메서드를 가진다 (유용하지는 않지만 반드시 이해해야함)
        * Object 타입의 인스턴스는 Object 타입의 프로퍼티와 메서드를 전부 상속한다.
            * constructor - 해당 객체를 만드는데 쓰인함수, Object() 함수가 생성자
            * hasOwnProperty( propertyName ): 해당 프로퍼티가 객체 인스턴스에 고유하며 프로토 타입에서 상속하지 않았음을 
                                              확인한다. 프로퍼티 이름은 반드시 문자열이어야한다. 
            * o.hasOwnProperty("name")
            * isPrototypeOf(object): 해당객체가 다른 객체의 프로토 타입인지 확인 
            * propertyIsEnumerable( propertyName ): 해당 프로퍼티를 for-in 문에서 나열할수 있는지 확인한다. 
                                                    프로퍼티 이름은 반드시 문자열!
            * toLocaleString(): 객체를 지역에 맞게 표현한 문자열 반환
            * toString() : 객체를 문자열로 변환해 반환한다.
            * valueOf() : 객체를 나타내는 문자열이나 숫자, 불리언을 반환한다. toString()과 같은 값을 반환할때가 많다.


## 3.6 연산자
* 단항연산자
     * 증감연산자: 정수에만 쓸 수 있는것이 아니며 문자열, 불리언, 부동소수점 숫자, 객체에도 쓸 수 있다.
        ```javascript
        // ex1
        var age = 29; 
        ++age; // 30

        // ex2
        var age = 29; 
        age = age + 1; // 30

        // ex3
        var age = 29;
        --age; // 28

        // ex4
        var age = 29;
        var anotherAge = --age + 2; // age: 28 저장 

        alert(age); // 28
        alert(anotherAge); // 30

        // ex5
        var num1 = 2;
        var num2 = 20;
        var num3 = --num1 + num2; // 21 (num1: 1저장)
        var num4 = num1 + num2; // 21
        ```
            
  * 증감연산자 규칙 
    * 유효한 숫자 형태의 문자열에 적용하면 해당 문자열을 숫자로 바꾼 후 증감한다
    ```javascript
        var s1 = "2";
        s1++; // 숫자3 (문자열에서 숫자로 바뀜)
    ```
    * 유효한 숫자 형태의 문자열이 아니라면 변수의 값에 NaN이 저장됨 
    ```javascript
        var s2 = "z";
        s2++ ; //NaN
    ```
    * 값이 false인 불리언에 적용하면 -> 0 으로 바꾼 후 증감 (불리언->숫자)
    ```javascript
        var b = false;
        b++ ; // 숫자 1
    ```
    * 값이 true -> 1로 바꾼 후 증감

    * 부동소수점 숫자에 적용하면 그대로 증감 
    ```javascript
        var f = 1.1;
        f--; // 0.10000000000009 (부동소수점의 부정확성 때문에)
    ```
    * 객체에 적용하면 해당 객체의 valueOf() 메서드를 먼저 호출해서 증감을 적용할 값을 얻는다.
    ```javascript
        var o = {
            valueOf: function() {
                return -1;
            }
        };
        o--; // 숫자 -2
    ```
     * 단항 플러스 연산자
        * 단항 플러스는 + 기호를 변수 앞에 써서 나타내며 숫자형 값에는 아무 효력이 없다.
            ```javascript
            var num = 25;
            num = +num; //여전히 25   
            ```
        * 숫자가 아닌 값에 적용하면 형변환 함수 Number()와 마찬가지로 동작함 
        * 불리언 값 false는 0 true 는 1 로 바뀐다.
        ```javascript
            // ex1
            var s1 = "01";
            s1 = +s1; // 숫자 1
            
            // ex2
            var s2 = "1.1";
            s2 = +s2; // 숫자 1.1

            // ex3
            var s3 = "z";
            s3 = +s3; // NaN

            // ex4
            var b = false;
            b = +b; //숫자 0
            
            // ex5
            var f = 1.1; 
            f = +f; // 변함없이 1.1

            // ex6
            var o = {
                valueOf: function() {
                    return -1;
                }
            };
            o = +o; // 숫자 -1
        ```
    * 단항 마이너스 연산자 : 보통 숫자형 값의 부호를 바꾸는 용도로 사용
        * 숫자가 아닌 값에 단항 마이너스를 적용하면 다음 규칙 적용하고 그 결과의 부호를 바꿈
        ```javascript
            // ex1
            var s1 = "01";
            s1 = -s1; // 숫자 -1

            // ex2
            var s2 = "1.1";
            s2 = -s2; // 숫자 -1.1

            // ex3
            var s3 = "z";
            s3 = -s3; // NaN

            // ex4
            var b = false;
            b = -b; //숫자 0

            // ex5
            var f = 1.1; 
            f = -f; // -1.1

            // ex6
            var o = {
                valueOf: function() {
                    return -1;
                }
            };
            o = -o; // 숫자 1  
          ```

    * 비트 연산자
        * 부호붙은 정수는 32비트중 첫 31비트를 써서 정수의 숫자형 값을 나타낸다. 32번째 비트가 숫자의 부호를 나타냄
        * 0은 양수, 1은 음수 
            * 숫자 18 -> 10010 -> 16+0+0+2+0
        * 음수도 이진 코드로 저장되지만 2의 보수라 불리는 형식을 사용함
            * 2의보수 구하는법: 절대값의 2진표현 구하기 -> 1의보수 구하기 (0을 1로, 1은 0으로 바꾸기) -> 결과에 1더하기
            ```javascript
            var num = -18;
            alert(num.toString(2)); // -10010
            ```
        * 비트 NOT 
        * 비트 AND
        * 비트 OR 
        * 비트 XOR

    * 불리언 연산자 
        * == 연산자 만큼이나 중요하며 if...else 반복문 쓸 때 중요
        * 논리 NOT: 값을 불리언 타입으로 바꾸는데 유용하다. Not 연산자를 연달아 두개쓰면 boolean() 함수를 쓴것과 마찬가지
            ```javascript
            alert(!false); // true
            alert(!"blue"); //false
            alert(!0); // true
            alert(!NaN); // true
            alert(!""); // true
            alert(!12345); //false
            * !! 두개쓰면 당연히 결과값은 반대
            ```
        * 논리 AND - 앰퍼센트 2개 (&&)로 나타낸다
            ```javascript
            var result = true && false;
            ```
            * 첫번째 피연산자에서 결과를 결정할 수 있다면 논리 AND 연산자는 두번째 피연산자를 결코 평가하지 않는다.
            * 논리 AND에서 첫번째 피연산자가 false라면 두번째 피연산자의 값과는 무관하게 결과는 true일 수 없다.
                ```javascript
                // ex1
                var found = true;
                var result = (found && someUndeclareVariable); // 에러발생 
                alert(result); //실행되지 않는다
                   
                // ex2
                var found = false; 
                var result = (found && someUndeclareVariable); // 에러없음
                alert(result); // 동작한다
                ```

        * 논리 OR
            ```javascript
            var result = true || false; 
            
            // ex1
            var found = true;
            var result = (found || someUndeclareVariable); // 에러없음 (true || false 정의되지 않은 변수)
            alert(result); // 동작한다.

            // ex2
            var found = false; 
            var result = (found || someUndeclareVariable); // 에러발생 
            alert(result); // 실행안됨
            ```
